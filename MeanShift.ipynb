```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x=iris.data
```

```python
from sklearn.cluster import MeanShift

ms = MeanShift()
ms.fit(x)
labels = ms.labels_
centers = ms.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,2],c=labels)
plt.scatter(centers[:,0],centers[:,2],marker='x',s=150,linewidths=5)
plt.show()
```
